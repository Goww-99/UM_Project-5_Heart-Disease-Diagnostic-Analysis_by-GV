#### *ETL (Extract, Transform, Load)*
ETL (Extract, Transform, Load) is a three-step process in computing used to manage data from various sources. The process begins with extracting data from one or more input sources, followed by transforming the data (including cleaning and formatting), and finally loading it into an output data container. ETL can be performed manually or through software applications that automate the process. Properly designed ETL systems ensure that the data meets specific standards and is ready for further use, such as in application development or decision-making. ETL is commonly used in data warehousing, where it integrates data from multiple systems, often managed by different stakeholders. The process involves extracting data, transforming it for storage and analysis, and loading it into a final database, such as a data warehouse, data lake, or data mart.

#### *What are the diagnostic tests for Heart Disease?*

There are several diagnostic tests for heart disease, including:-

1. **Electrocardiogram (ECG)**: A short, painless test that records the electrical activity of your heart.
2. **Echocardiogram**: An ultrasound of the heart that uses sound waves to examine it.
3. **Blood tests**: Can check for substances in your blood that may indicate heart failure or other illnesses.
4. **Cholesterol test**: Can help determine your cholesterol levels, which can be a risk factor for heart disease.
5. **Plasma ceramides test**: Measures the levels of ceramides in your blood, which can indicate a higher risk of heart disease.
6. **Exercise stress test**: Evaluates how your heart performs under physical stress.
7. **Thallium scan**: Injects a radioactive substance into a vein and uses a camera to create pictures of your heart.
8. **Blood pressure monitor**: Can be worn for 24 hours to help your doctor get accurate information about your blood pressure.

12. ca: Number of major vessels (0-3) colored by flourosopy.
